```python
df.info()
```

```python
df['hourly'] = df["Salary Estimate"].apply(lambda x: 1 if 'per hour' in x.lower() else 0 )
df['employer provided'] = df["Salary Estimate"].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0 )
df
```

```python
#removing invalid salaries
df = df[df["Salary Estimate"] != "-1"]
#split text and remove unecessary texts
salary = df["Salary Estimate"].apply(lambda x: x.split("(")[0])
ex_kd = salary.apply(lambda x: x.replace("K", "").replace('$', ''))
ex_rem_text = ex_kd.apply(lambda x: x.lower().replace('per hour', '').replace('employer provided salary:', ''))
ex_rem_text
```

```python
df['min salary'] = ex_rem_text.apply(lambda x: int(x.split('-')[0]))
df['max salary'] = ex_rem_text.apply(lambda x: int(x.split('-')[1]))
df['average salary'] = (df['min salary'] + df['max salary'])/2
df
```

```python
#clean company text
df['Company'] = df.apply(lambda x: x['Company Name'] if x['Rating']< 0 else x['Company Name'][:-3], axis=1)
df['Company'] = df['Company'].apply(lambda x: x.replace('\n', ''))
df
```

```python
#states
df['State'] = df['Location'].apply(lambda x: x.split(',')[1])
df.State.value_counts()
```

```python
df['Same State'] = df.apply(lambda x: 1 if x['Location'] == x['Headquarters'] else 0, axis=1)
df
```

```python
#co age
df['Age'] = df['Founded'].apply(lambda x: x if x < 1 else 2025 - x)
df
```

```python
#python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
 
#r studio 
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df.R_yn.value_counts()

#spark 
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df.spark.value_counts()

#aws 
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df.aws.value_counts()

#excel
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df.excel.value_counts()
```

```python
df_out = df.drop(['Unnamed: 0'], axis =1)
df_out.to_csv('salary_data_cleaned.csv',index = False)
```
